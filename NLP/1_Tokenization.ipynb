```python
corpus = """Hello, Welcome to my world. My name is Deepak kumar yati, and I have completed Python, ML, DL, Gen and Agentic AI! and I am revising these topics again.

"""
```

```python
print(corpus)
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
## Tokenization
## Paragraph -> Sentences

from nltk.tokenize import sent_tokenize
```

```python
sent_tokenize(corpus)
```

```python
documents=sent_tokenize(corpus)

type(documents)
```

```python
for sentence in documents:
    print(sentence)
```

```python
## Tokenization
## Paragraph -->words
## Sentence --->words


from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
# sentence --> words

for sentence in documents:
    print(word_tokenize(sentence))
```

```python
# punctuation considered as separate words

from nltk.tokenize import wordpunct_tokenize

wordpunct_tokenize(corpus)
```

```python
#with this '.' is not treted as separate word

from nltk.tokenize import TreebankWordTokenizer

tokenizer=TreebankWordTokenizer()

tokenizer.tokenize(corpus)
```
