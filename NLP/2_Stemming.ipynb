# Stemming

```python
## classification problem
## customer review - positive, negative
## eating, eaten --> eat
## going, gone, goes  --> go
```

```python
words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
```

# PorterStemmer

```python
from nltk.stem import PorterStemmer
```

```python
stemming=PorterStemmer()
```

```python
for word in words:
    print(word+"----->"+stemming.stem(word))
```

```python
stemming.stem('playing')
```

```python
stemming.stem('congratulations') #major disadvantage, changing the form of word
```

```python
stemming.stem('sitting')
```

# RegexpStemmer class

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer = RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('goes')
```

```python
reg_stemmer.stem('plays')
```

```python
reg_stemmer.stem('understandable')
```

# Snowball Stemmer

```python
## gives better form of words than PorterStemmer

from nltk.stem import SnowballStemmer
```

```python
snowballstemmer= SnowballStemmer('english')
```

```python
for word in words:
    print(word+"------->"+snowballstemmer.stem(word))
```

```python
snowballstemmer.stem('congratulations')
```

```python
#Porter Stemmer
print(stemming.stem('fairly'))


#snowball Stemmer
print(snowballstemmer.stem('fairly'))
```
